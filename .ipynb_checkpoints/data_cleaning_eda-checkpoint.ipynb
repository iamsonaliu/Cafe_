```python
import pandas as pd
```

```python
df=pd.read_csv("cafes_ddn_hrd_rishikesh_mussoorie.csv")
df
```

```python
print("Shape:", df.shape)  
print("\nColumn Info:")
print(df.info())
df.head(3)
```

```python
# Missing value count
df.isnull().sum()
# Percentage of missing
(df.isnull().sum()/len(df))*100
```

```python
df['CUSINE_CATEGORY'] = df['CUSINE_CATEGORY'].fillna("Unknown")
```

```python
df.head()
```

```python
df['RATING'] = pd.to_numeric(df['RATING'], errors='coerce')
```

```python
def get_rating_type(r):
    if pd.isna(r): 
        return "Unknown"
    elif r >= 4.5: 
        return "Excellent"
    elif r >= 4.0: 
        return "Very Good"
    elif r >= 3.0: 
        return "Average"
    else: 
        return "Poor"
df['RATING_TYPE'] = df['RATING'].apply(get_rating_type)
df.head(5)
```

```python
print("Duplicates:", df.duplicated().sum())
```

```python
print("Unique cities:", df['CITY'].nunique())
print("Unique cuisines:", df['CUSINE_CATEGORY'].nunique())
print("Unique dining types:", df['CUSINE TYPE'].nunique())

df.groupby('CITY')['PRICE'].mean()
df.groupby('CITY')['RATING'].mean()

```

```python
df.tail()
```

```python
print(df.size)   # total number of elements (rows × columns)
print(df.shape)  # (rows, columns)
print(len(df))   # number of rows
```

```python
print(df.info())
```

```python
df["VOTES"] = pd.to_numeric(df["VOTES"], errors="coerce")
```

```python
df
```

```python
df['HAS_RATING'] = df['RATING'].notna().astype(int)
df
```

```python
df['RATING'] = df.groupby('CITY')['RATING'].transform(
    lambda x: x.fillna(x.mean())
)
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# =============================
# 1. Correlation Heatmap (Numerical Features)
# =============================
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Numerical Features")
plt.show()

# =============================
# 2. Distribution of Cost vs Rating
# =============================
plt.figure(figsize=(8,6))
sns.scatterplot(x="PRICE", y="RATING", data=df, alpha=0.6)
plt.title("Price vs Rating")
plt.show()

# =============================
# 3. Cuisines vs Average Rating
# =============================
top_cuisines = df.groupby("CUSINE TYPE")["RATING"].mean().sort_values(ascending=False).head(10)
print("Top 10 Cuisine Types with Highest Average Ratings:")
print(top_cuisines)

plt.figure(figsize=(10,5))
sns.barplot(x=top_cuisines.index, y=top_cuisines.values)
plt.xticks(rotation=45, ha="right")
plt.title("Top 10 CuisineTypes with Highest Average Ratings")
plt.show()

top_cuisine_cat = df.groupby("CUSINE_CATEGORY")["RATING"].mean().sort_values(ascending=False).head(10)
print("Top 10 Cuisine Category with Highest Average Ratings:")
print(top_cuisine_cat)

plt.figure(figsize=(10,5))
sns.barplot(x=top_cuisine_cat.index, y=top_cuisine_cat.values)
plt.xticks(rotation=45, ha="right")
plt.title("Top 10 Cuisine Categories with Highest Average Ratings")
plt.show()

# =============================
# 4. Votes vs Rating
# =============================
plt.figure(figsize=(8,6))
sns.scatterplot(x="VOTES", y="RATING", data=df, alpha=0.6)
plt.title("Votes vs Rating")
plt.show()

# =============================
# 5. Count of Rating Types
# =============================
plt.figure(figsize=(6,4))
sns.countplot(x="RATING_TYPE", data=df, order=df["RATING_TYPE"].value_counts().index)
plt.title("Count of Each Rating Type")
plt.show()
```
