```python
print("Shape:", df.shape)  
print("\nColumn Info:")
print(df.info())
df.head(3)
```

```python
# Missing value count
df.isnull().sum()
# Percentage of missing
(df.isnull().sum()/len(df))*100
```

```python
df['CUSINE_CATEGORY'] = df['CUSINE_CATEGORY'].fillna("Unknown")
```

```python
df.head()
```

```python
df['RATING'] = pd.to_numeric(df['RATING'], errors='coerce')
```

```python
def get_rating_type(r):
    if pd.isna(r): 
        return "Unknown"
    elif r >= 4.5: 
        return "Excellent"
    elif r >= 4.0: 
        return "Very Good"
    elif r >= 3.0: 
        return "Average"
    else: 
        return "Poor"
df['RATING_TYPE'] = df['RATING'].apply(get_rating_type)
df.head(5)
```
