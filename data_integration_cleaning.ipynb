```python
import pandas as pd
import glob
import os
```

```python
path_dehradun = "Dehradun"
all_files = glob.glob(os.path.join(path_dehradun, "*.csv"))
df_list = [pd.read_csv(f,sep='|') for f in all_files]
df_dehradun = pd.concat(df_list, ignore_index=True)
df_dehradun["city"] = "Dehradun"
```

```python
path_haridwar = "Haridwar"
all_files = glob.glob(os.path.join(path_haridwar, "*.csv"))
df_list = [pd.read_csv(f,sep='|') for f in all_files]
df_haridwar = pd.concat(df_list, ignore_index=True)
df_haridwar["city"] = "Haridwar"
```

```python
df_mussoorie = pd.read_csv("Mussoorie/1-Mussooriehotels.csv",sep='|')
df_mussoorie["city"] = "Mussoorie"

df_rishikesh = pd.read_csv("Rishikesh/1-Rishikeshhotels.csv",sep='|')
df_rishikesh["city"] = "Rishikesh"
```

```python
df=pd.concat([df_dehradun,df_haridwar,df_mussoorie,df_rishikesh])
df
```

```python
df.to_csv("cafes_ddn_hrd_rishikesh_mussoorie.csv", index=False)
```
